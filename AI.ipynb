## Entrega com Bot/IA para nosso Projeto do Challenge, Novas melhorisa virão em próximas entregas.


```python
import pyttsx3, speech_recognition as sr

engine = pyttsx3.init()

voices = engine.getProperty("voices") 
engine.setProperty('voice', voices[0].id) #brasil
engine.setProperty('volume', 1)
engine.setProperty("rate", 155)


engine.runAndWait() 

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

# Ouve o usuario.
def command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("\n  Listening...\n")
        speak('ouvindo!')
        r.adjust_for_ambient_noise = 1.25
        r.pause_threshold = 1
        audio = r.listen(source)
    try:
        print("  Recognizing...\n")
        speak('Espera um pouco pra eu tentar entender.')
        query = r.recognize_google(audio, language='pt')
        #print(query)
    except:
        return "©empty_^_^_queryª"
    return query




speak("Olá, sou o seu assistente de Compras! Posso te ajudar em algo?  Por favor, responda com Sim ou Não")


if __name__ == "__main__":
        
    while True:
        query = command().lower()
        if 'sim' in query or 'Sim'in query or 'sin' in query:
            print("SIM")
            speak("certo, Tenho algumas opções em que posso te ajudar, como opção um ,achar um produto ou opção 2, verificar o preço de um produto")  
            while True:
                query = command().lower()   
                    
                if '1' in query or 'um' in query:
                    
                    
                    print("escolha: 1 ")
                    speak("você escolheu a opção numero 1")
                    speak("Tenho algumas opções de produtos para te ajudar a encontrar, como frutas , carnes e laticínios. Escolha um deles!, ou se quiser escolher outra ajuda, diga SAIR")
                    
                    while True:
                        
                        query = command().lower()
                        
                        if 'frutas' in query or 'fruta' in query:
                            speak("As frutas estão após o corredor 5, próximo as geladeiras de frios.")
                        elif 'carnes' in query or 'carne' in query or 'CARNE' in query:
                            speak("As carnes podem ser encontradas, no ultimo corredor do mercado, de número 8 na seção de açogue")
                        elif 'laticinio' in query or 'laticínios' in query or "LATICÍNIOS" in query:
                            speak("os laticínios estão no início do corredor 7, próximos das bebidas.") 
                    
                        
                        elif 'fim' in query or 'sair' in query : 
                            print("sair")
                            speak("Voltando ao Menu das opções de ajuda")
                            break
                    
                    
                        else:
                            print("opção invalida 3")
                            speak('não entendi o que você disse.')
                            speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
                
                elif '2' in query or 'dois' in query:
                    print("escolha : 2")
                    speak('você escolheu a opção numero 2')
                    speak('Bem, por enquanto, posso te dizer o preço de alguns produtos, como o requeijão, café, ou leite, escolha um! e não se preocupe, novas funções virão futuramente')
                    while True:
                        
                        query = command().lower()
                        
                        if 'requeijão' in query or 'requeijao' in query:
                            speak("o preço do requeijão é de 7 reais o pode pequeno e 15 reais o pote grande")
                        elif 'café' in query or 'cafe' in query or 'CAFE' in query:
                            speak("O preço do café subiu um pouco e hoje meio quilo está por 15 reais")
                        elif 'leite' in query or 'LEITE' in query or "Leites" in query:
                            speak("O litro de leite varia de 4 a 5 reais dependendo da marca.") 
                    
                        
                        elif 'fim' in query or 'sair' in query : 
                            print("sair")
                            speak("Voltando ao Menu das opções de ajuda")
                            break
                    
                    
                        else:
                            print("opção invalida 3")
                            speak('não entendi o que você disse.')
                            speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
                
                elif 'fim' in query or 'sair' in query : 
                    print("sair")
                    speak("Voltando ao Menu do Sim e não")
                    break
                            
                else:
                    print("opção invalida 2")
                    speak('não entendi o que você disse.')
                    speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
            
        elif 'não' in query or 'Não' in query:
            print("opção: 2")
            speak("Está bem, até logo!")
            break
                
                  
        else:
            print("opção invalida 1")
            speak('não entendi o que você disse.')
            speak("Por favor, vamos tentar novamente. escolha uma opção válida.")
            
        
        
```
